```python
#Importing the libraries
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
from sklearn.cluster import KMeans
```

```python
# load the raccoon face as a numpy array
from scipy.misc import face
face = face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
```

```python
face.shape
```

```python
# # We need an (n_sample, n_feature) array
img_r = face.reshape((-1, 1))
```

```python
img_r.shape
```

```python
# Applying KMeans Clustering
k_gray = cluster.KMeans(n_clusters=5).fit(img_r)
y_pred = k_gray.predict(img_r)
```

```python
values = k_gray.cluster_centers_
```

```python
labels = k_gray.labels_
```

```python
img_compressed = np.choose(labels, values)
```

```python
img_compressed=np.reshape(img_compressed, (face.shape))
```

```python
## Visualising the Result
fig = plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(face, cmap=plt.cm.gray)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Compressed Image using K-Means')
ax.imshow(img_compressed,cmap=plt.cm.gray)
plt.show()
```
